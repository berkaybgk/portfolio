```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('resources/kc_house_data.csv')
```

```python
# Check for missing values
df.isnull().sum()
```

This data set is already clean and does not have any missing values.

It was created to play around and practice with the data set. It even has engineered features like `sqft_living15` and `sqft_lot15` which are the average square footage of the 15 closest neighbors. This makes our job easier as we don't have to do any feature engineering.


```python
# See if there is anything useful with the zipcode
df['zipcode'].unique()
```

We can try to use encoding to preserve information from the zipcode but it would be highly sparse. We can instead take the 3rd digit of the zipcode to get the area code.

1 - Urban

2 - Suburban

So it should give a rough idea.


```python
def get_area_code(zipcode):
    return int(str(zipcode)[2])

df['area_code'] = df['zipcode'].apply(get_area_code)

# Drop the zipcode column
df.drop('zipcode', axis=1, inplace=True)
```

```python
# Visualize the lat and long with respect to price
# This will give us an idea of the location of the houses, like on a map

plt.figure(figsize=(10, 6))
sns.scatterplot(x='long', y='lat', data=df, hue='price', palette='inferno', alpha=0.5, s=16)
plt.title('Latitude and Longitude vs Price')
```

We see a map shape occuring in the data. This is because the data is from King County, Washington. We can see that the houses are more expensive near the water.


```python
numeric_cols = df.select_dtypes(include=[np.number]).columns

df_copy = df.copy()
df_copy = df_copy[numeric_cols]

# Create a heatmap to see the correlation between the features
plt.figure(figsize=(12, 8))
sns.heatmap(df_copy.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')

```

We see high association between the following features:

- `sqft_living` and `price`
- `sqft_above` and `sqft_living`
- `sqft_living` and `grade`
- `sqft_living` and `bathrooms`

The price is highly correlated with the square footage of the living area. This is expected as the price of a house is usually determined by the size of the house.

We can also see that the `yr_built` is inversely correlated with the `area_code`, which causes because new houses are built in the expanding suburban areas.
So, new places, new houses.


```python
# Plot the sqft_living vs price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sqft_living', y='price', data=df, alpha=0.5, color='purple')
plt.title('Square Footage of Living Area vs Price')
```

```python
# See if locality actually exists, compare sqft_living with sqft_living15
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sqft_living', y='sqft_living15', data=df, alpha=0.5, color='green')
plt.title('Square Footage of Living Area vs Square Footage of Living Area of 15 Neighbors')
```

It looks like there is a positive correlation between the square footage of the living area and the square footage of the living area of the 15 closest neighbors.

This is expected as the houses in the same area are usually of the same size.


```python
# See the renovated houses' renovation year and built year

renovated_houses = df[df['yr_renovated'] != 0]

plt.figure(figsize=(10, 6))
sns.scatterplot(x='yr_built', y='yr_renovated', data=renovated_houses, alpha=0.5, color='red')
plt.title('Year Built vs Year Renovated')
```

Most of the renovated houses were built in the early 1900s. 

This is expected as the houses built back then would need renovation.

```python
# See the grade distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='grade', data=df)
plt.title('Grade Distribution')
```
