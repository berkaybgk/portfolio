```python
df["Status"] = df["Status"].apply(lambda x: 0 if x == "Developing" else 1)
```

```python
df["Year"] = df["Year"].apply(lambda x: x-2006 + 0.0)
```

```python
df.info()
```

```python
def fill_na_by_country_mean(df, column_to_fill):
    df = df.copy()

    # Compute the mean for each country
    country_means = df.groupby("Country")[column_to_fill].transform("mean")

    # Fill NaN values with the computed means
    df[column_to_fill] = df[column_to_fill].fillna(country_means)

    return df

def clean_life_exp_data(df: pd.DataFrame):
    df = df.copy()

    try:
        # Drop rows with missing target values
        df = df.dropna(subset=["Life expectancy"])

        # First fill by country mean
        for feature in df.columns[df.isnull().any()]:
            df = fill_na_by_country_mean(df, feature)

        # Some countries don't have some type of data, use global mean
        for feature in df.columns[df.isnull().any()]:
            df[feature] = df[feature].fillna(df[feature].mean())

        # Convert columns to float
        df["under-five deaths"] = df["under-five deaths"].apply(lambda x: x+0.0)
        df["Measles"] = df["Measles"].apply(lambda x: x+0.0)
        df["infant deaths"] = df["infant deaths"].apply(lambda x: x+0.0)
        df["Status"] = df["Status"].apply(lambda x: x+0.0)

        # Since the country features are already represented
        df.drop(columns=["Country"], inplace=True)

        return df

    except:
        print("The columns might already be dropped")
        return df

```

```python
df = clean_life_exp_data(df)
df.info()
```

```python
df.to_csv("resources/cleaned_life_expectancy.csv")
```
